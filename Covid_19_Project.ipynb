# Making parent table by merging delta,delta21_14,delta7,meta and total.

Rename column name.

```python
delta.rename(columns = {'tested':'tested_delta','vaccinated1':'vaccinated1_delta','vaccinated2':'vaccinated2_delta','confirmed':'confirmed_delta','deceased':'deceased_delta','recovered':'recovered_delta','other':'other_delta'},inplace=True)
```

```python
delta21_14.rename(columns = {'confirmed':'confirmed_delta21_14'},inplace=True)
```

```python
delta7.rename(columns = {'tested':'tested_delta7','vaccinated1':'vaccinated1_delta7','vaccinated2':'vaccinated2_delta7','confirmed':'confirmed_delta7','deceased':'deceased_delta7','recovered':'recovered_delta7','other':'other_delta7'},inplace=True)
```

```python
meta.rename(columns = {'date':'date_meta','last_updated':'last_updated_meta','population':'population_meta','tested.date':'tested.date_meta','tested.source':'tested.source_meta','vaccinated.date':'vaccinated.date_meta','notes':'notes_meta'},inplace=True)
```

```python
total.rename(columns = {'tested':'tested_total','vaccinated1':'vaccinated1_total','vaccinated2':'vaccinated2_total','confirmed':'confirmed_total','deceased':'deceased_total','recovered':'recovered_total','other':'other_total'},inplace=True)
```

Merge table on state_code column as it is common for all .

```python
parent_table = pd.merge(pd.merge(pd.merge(pd.merge(delta,delta21_14,on = 'state_code',how='outer'),delta7,on = 'state_code',how='outer'),meta,on = 'state_code',how='outer'),total,on = 'state_code',how='outer')
```
